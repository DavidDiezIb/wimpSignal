```python
## Poisson method. En general casi un orden de magnitud por encima la curva de sensibilidad
from scipy.integrate import quad
mass = [ 0.1, 0.15, 0.20, 0.22, 0.25, 0.28, 0.3, 0.33, 0.35, 0.4, 0.45, 0.5, 0.6, 0.7, 0.8, 0.85, 0.9, 0.95, 1, 1.05, 1.1, 1.15, 1.2, 1.3, 1.5, 1.7, 1.9, 2, 2.3, 2.6, 3, 3.5, 4, 5, 6, 7, 8, 9, 10, 15, 20, 30, 50, 70, 100, 1000, 10000 ]
area = np.zeros(np.size(mass))
areaRThr = np.zeros(np.size(mass))
areaN = np.zeros(np.size(mass))
areaC = np.zeros(np.size(mass))
areaH = np.zeros(np.size(mass))
areaX1T = np.zeros(np.size(mass))
areaAalbers = np.zeros(np.size(mass))
areaAalbersH = np.zeros(np.size(mass))
areaAalbersX1T = np.zeros(np.size(mass))
crossLim = np.zeros(np.size(mass))
crossLimRThr = np.zeros(np.size(mass))
crossLimN = np.zeros(np.size(mass))
crossLimC = np.zeros(np.size(mass))
crossLimH = np.zeros(np.size(mass))
crossLimX1T = np.zeros(np.size(mass))
crossLimAA = np.zeros(np.size(mass))
crossLimAAH = np.zeros(np.size(mass))
crossLimAAX1T = np.zeros(np.size(mass))

thr = 0.6 # Threshold (keV nuclear recoil)
thrR = 0.6 # Threshold reduced (keV nuclear recoil)
binW = 1.0 # Bin width (keV)
binWR = 0.1 # Bin width (keV)

nsig = wr.survivalPoiss(0.1, 109.5*binW) # Model B 109.5 signals in 0.6-1.6 keV. Xenon1T 755 bkg events in 4.9-40.9 keV
nsigX = wr.survivalPoiss(0.1, 755) # Xenon1T 755 bkg events in 4.9-40.9 keV
nsigR = wr.survivalPoiss(0.1, 109.5*binWR) # Model B 109.5/2 signals in 0.6-1.1 keV.
print(nsig)
print(nsigR)

for i in range(0, np.size(mass)):
    area[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, A, T, MD), thr, thr+binW)[0]*2*0.16*10**(-3) # Rate x Exposure
    areaRThr[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, A, T, MD), thrR, thrR+binWR)[0]*2*0.16*10**(-3) # Rate x Exposure. Smaller bin
    areaAalbers[i]=quad(lambda e:  wimprates.rate_wimp_std(e, mass[i], sigma_nucleon=1e-45, material='Ne'), thr, thr+binW)[0]*0.32*10**(-3)
    areaX1T[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, A, T, MD), 4.9, 40.9)[0]*1
    areaAalbersX1T[i]=quad(lambda e:  wimprates.rate_wimp_std(e, mass[i], sigma_nucleon=1e-45, material='Xe'), 4.9, 40.9)[0]*1

    areaN[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, ATOMIC_WEIGHT['Ne'], T, MD), thr, thr+binW)[0]*2*0.15955*10**(-3) # Rate x Exposure. 98% Neon
    areaC[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, ATOMIC_WEIGHT['C'], T, MD), thr, thr+binW)[0]*2*0.00775*10**(-3) # Rate x Exposure. 2% isobutano C4H10
    areaH[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, ATOMIC_WEIGHT['H'], T, MD), thr, thr+binW)[0]*2*0.00161*10**(-3) # Rate x Exposure. 2% isobutano
    areaAalbersH[i]=quad(lambda e:  wimprates.rate_wimp_std(e, mass[i], sigma_nucleon=1e-45, material='H'), thr, thr+binW)[0]*2*0.00161*10**(-3)

    crossLim[i] = 10**(-45)*nsig/area[i]
    crossLimRThr[i] = 10**(-45)*nsigR/areaRThr[i]
    crossLimAA[i] = 10**(-45)*nsig/areaAalbers[i]
    crossLimX1T[i] = 10**(-45)*nsigX/areaX1T[i]
    crossLimAAX1T[i] = 10**(-45)*nsigX/areaAalbersX1T[i]

    crossLimN[i] = 10**(-45)*nsig/areaN[i]
    crossLimC[i] = 10**(-45)*nsig/areaC[i]
    crossLimH[i] = 10**(-45)*nsig/areaH[i]
    crossLimAAH[i] = 10**(-45)*nsig/areaAalbersH[i]
```
