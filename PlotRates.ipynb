```python
## Poisson method. En general casi un orden de magnitud por encima la curva de sensibilidad
from scipy.integrate import quad
mass = [ 0.075, 0.1, 0.12, 0.15, 0.17, 0.20, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.6, 0.7, 0.8, 0.85, 0.9, 0.95, 1, 1.2, 1.3, 1.5, 1.7, 1.9, 2, 2.3, 2.6, 3, 3.5, 4, 5, 6, 7, 8, 9, 10, 15, 20, 30, 50, 70, 100, 1000, 10000 ]
area = np.zeros(np.size(mass))
areaN = np.zeros(np.size(mass))
areaC = np.zeros(np.size(mass))
areaH = np.zeros(np.size(mass))
areaX1T = np.zeros(np.size(mass))
areaAalbers = np.zeros(np.size(mass))
areaAalbersH = np.zeros(np.size(mass))
areaAalbersX1T = np.zeros(np.size(mass))
crossLim = np.zeros(np.size(mass))
crossLimN = np.zeros(np.size(mass))
crossLimC = np.zeros(np.size(mass))
crossLimH = np.zeros(np.size(mass))
crossLimX1T = np.zeros(np.size(mass))
crossLimAA = np.zeros(np.size(mass))
crossLimAAH = np.zeros(np.size(mass))
crossLimAAX1T = np.zeros(np.size(mass))

nsig = wr.survivalPoiss(0.1, 109.5) # Model B 109.5 signals in 0.6-1.6 keV. Xenon1T 755 bkg events in 4.9-40.9 keV
#print(nsig)

for i in range(0, np.size(mass)):
    area[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, A, T, MD), 0.6, 1.6)[0]*2*0.16*10**(-3) # Rate x Exposure
    areaAalbers[i]=quad(lambda e:  wimprates.rate_wimp_std(e, mass[i], sigma_nucleon=1e-45, material='Ne'), 0.6, 1.6)[0]*0.32*10**(-3)
    areaX1T[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, A, T, MD), 4.9, 40.9)[0]*1
    areaAalbersX1T[i]=quad(lambda e:  wimprates.rate_wimp_std(e, mass[i], sigma_nucleon=1e-45, material='Xe'), 4.9, 40.9)[0]*1

    areaN[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, ATOMIC_WEIGHT['Ne'], T, MD), 0.6, 1.6)[0]*2*0.15955*10**(-3) # Rate x Exposure. 98% Neon
    areaC[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, ATOMIC_WEIGHT['C'], T, MD), 0.6, 1.6)[0]*2*0.00775*10**(-3) # Rate x Exposure. 2% isobutano C4H10
    areaH[i]=quad(lambda e: wr.rate(e, mass[i], sigma0, ATOMIC_WEIGHT['H'], T, MD), 0.6, 1.6)[0]*2*0.00161*10**(-3) # Rate x Exposure. 2% isobutano
    areaAalbersH[i]=quad(lambda e:  wimprates.rate_wimp_std(e, mass[i], sigma_nucleon=1e-45, material='H'), 0.6, 1.6)[0]*2*0.00161*10**(-3)

    crossLim[i] = 10**(-45)*nsig/area[i]
    crossLimAA[i] = 10**(-45)*nsig/areaAalbers[i]
    crossLimX1T[i] = 10**(-45)*nsig/areaX1T[i]
    crossLimAAX1T[i] = 10**(-45)*nsig/areaAalbersX1T[i]

    crossLimN[i] = 10**(-45)*nsig/areaN[i]
    crossLimC[i] = 10**(-45)*nsig/areaC[i]
    crossLimH[i] = 10**(-45)*nsig/areaH[i]
    crossLimAAH[i] = 10**(-45)*nsig/areaAalbersH[i]
```
